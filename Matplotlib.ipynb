# plt.plot Fonksiyonu

```python
# plt.plot() fonksiyonu, Matplotlib kütüphanesinde temel çizgi grafiği çizmek için kullanılır.
#İki liste veya dizi arasındaki ilişkiyi göstermek için x ve y eksenlerinde değerleri birleştiren düz çizgiler çizer.

# Fonksiyonun kullanımı oldukça basittir. İki eşit uzunluktaki liste veya dizi alır ve bunları x ve y eksenlerinde değerler olarak
# kullanarak bir çizgi grafiği oluşturur.

# Örnek olarak, iki liste arasındaki ilişkiyi gösteren basit bir çizgi grafiği oluşturalım:
```

## plt.plot ve plt.scatter Parametreleri

### plt.plot:

color: çizginin rengini belirler

linestyle: çizgi stilini belirler

marker: noktalara denk gelen kesişimleri gösterir. "o" olabilir, "s" olabilir.

markersize:

alpha: saydamlık - opaklık. 1=opak, 0=saydam

label: grafiğin etiketi

```python
# x = df["area"]
# y = df["price"]
x = np.arange(1,11)
y = np.arange(1,11) * 2
plt.figure(figsize=(8,6))
plt.plot(x,y, "--", color = "green")
# plt.plot(x,y, "g--")
plt.show()
```

# plt.scatter

```python
# plt.scatter() fonksiyonu, Matplotlib kütüphanesinde nokta grafiği (scatter plot) çizmek için kullanılır.
# Nokta grafiği, iki değişken arasındaki ilişkiyi göstermek için x ve y eksenlerinde noktaların dağılımını çizerek görselleştirir.
# Bu tür grafikler, iki değişken arasında doğru veya ters ilişkiler gibi eğilimleri ve kümelenmeleri görmeye yardımcı olabilir.
```

##plt.scatter parametreleri

color: renk belirtir

label: grafiğin etiketi

s: noktaların büyüklüğü

edgecolors: noktaları saran renk.

alpha: saydamlık ve opaklık

```python
# modellediğimiz ilişki - fonksiyon = f(x) = 2x
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 2, 4, 6, 8, 10, 12]

plt.figure(figsize = (6, 4))

# Aynı anda plt.plot() ve plt.scatter() fonksiyonlarını kullanarak çizgi grafiği (plt.plot)
# ve nokta grafiğini (scatter plot) aynı figure üzerinde çizdirebilirsiniz:
plt.scatter(x, y, color = "red")
plt.plot(x, y, color="blue")

plt.show()
```

```python
x = np.arange(100) / 10
y1 = np.sin(x)
y2 = np.cos(x)

plt.scatter(x, y2, color="red", label = "Cosx")
plt.plot(x, y1, color = "green", label = "Sinx")

plt.legend(loc="upper right")
plt.show()
```

```python
plt.plot(x, y1, color = "green", label = "Sinx",
         linestyle="--", )

plt.legend(loc="upper right")
plt.show()
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 2, 4, 6, 8, 10, 12]

plt.figure(figsize = (6, 4))

plt.plot(x, y, color="blue", marker="s",
         markersize=10, alpha=0.5)

plt.show()
```

```python
x = np.arange(100) / 10
plt.scatter(x, np.cos(x), color="red", label = "Cosx",
            s=40, edgecolors="blue", linewidths=2,
            alpha=0.5)

plt.legend()
plt.show()
```

# plt.bar()

```python
# plt.bar() bir sütun grafiği çizer. Bu grafik türü, kategorilerin değerleri arasındaki ilişkiyi gösterir.
```

## Parametreler
color: Sütunların rengini belirtir.

width: Sütunların genişliğini kontrol eder

edgecolor: Sütunların kenarlık rengini belirler

linewidth: Kenarlık kalınlığını belirtir.

```python
x = ["A", "B", "C", "D"]
y = [5, 10, 14, 18 ]

plt.figure(figsize=(6,4))
plt.bar(x, y, color = "red", width = 0.4, edgecolor= "black", linewidth= 2)
plt.show()
```

# plt.hist()

```python
# plt.hist() fonksiyonu, verilerin dağılımını göstermek için bir histogram grafiği çizer.
# Histogramlar, verilerin sıklığını veya yoğunluğunu görselleştirmek için kullanılır.
```

## Parametreler

bins: Grup sayısını belirler.

range: Bin'lerin alt ve üst limitini belirler.

density: True olursa grafik yoğunluk belirtir, false olursa frekans belirtir.

color: Bin rengi

alpha: Saydamlık opaklık

```python
plt.hist(df["bedrooms"])
plt.show()
```

```python
plt.hist(df["area"], bins= 20, alpha= 0.7, color= "red", label="area")
plt.title("Area Frequency Graph")
plt.xlabel("range")
plt.ylabel("frequency")
plt.legend()
plt.show()
```
